```python
import numpy as np
import random
```

```python
# fitness function
def fitness_fn(x):
  return x**2
num_particles = 100
max_iter = 100
c1 = 1.5
c2 = 1.75
w = 0.5
lower_bound = -10
upper_bound = 10

```

```python
position = np.random.uniform(lower_bound,upper_bound,num_particles)
velocity = np.random.uniform(-1,1,num_particles)
p_best_pos = position.copy()
p_best_val = np.array([fitness_fn(x) for x in position])

gbest_idx = np.argmin(p_best_val)
gbest_pos = p_best_pos[gbest_idx]
gbest_val = p_best_val[gbest_idx]
```

```python
for iter in range(max_iter):

    for i in range(num_particles):
        r1 = random.random()
        r2 = random.random()

        # velocity update
        velocity[i] = (
            w * velocity[i]
            + c1 * r1 * (p_best_pos[i] - position[i])
            + c2 * r2 * (gbest_pos - position[i])
        )

        # position update
        position[i] = position[i] + velocity[i]

        # fitness evaluation
        score = fitness_fn(position[i])

        # update personal best
        if score < p_best_val[i]:
            p_best_val[i] = score
            p_best_pos[i] = position[i]

    # ✅ update global best (OUTSIDE particle loop)
    best_idx = np.argmin(p_best_val)
    if p_best_val[best_idx] < gbest_val:
        gbest_val = p_best_val[best_idx]
        gbest_pos = p_best_pos[best_idx]

    # ✅ print once per iteration
    print(f"Iter {iter+1}: Best = {gbest_pos:.10f}")
```

```python
print("final best ",gbest_pos)
print("best score: ",gbest_val)
```
